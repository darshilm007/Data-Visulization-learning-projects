```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

data = {
    'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar', 'Apr', 'Apr'],
    'Region': ['North', 'South', 'North', 'South', 'North', 'South', 'North', 'South'],
    'Sales': [1200, 900, 1500, 1100, 1700, 1400, 1600, 1300],
    'Customers': [50, 40, 60, 48, 70, 60, 65, 55]
}

df = pd.DataFrame(data)


"""1 Monthly Sales Trend by Region

Line plot

X  Month

Y  Sales

Hue  Region"""

fig, axes = plt.subplots(1, 3, figsize=(18,6))
fig.suptitle("Sales Performance & Customer Insights",fontsize=16)

sns.lineplot(x='Month',y='Sales',data=df,hue='Region',ax=axes[0])
axes[0].set_title("Monthly Sales Trend")
axes[0].set_xlabel("Month")
axes[0].set_ylabel("Sales")

"""2 Sales vs Customers Relationship

Scatter plot

X  Customers

Y  Sales

Hue  Region"""

sns.scatterplot(x='Customers',y='Sales',data=df,hue='Region',ax=axes[1])
axes[1].set_title("Sales vs Customers Relationship")
axes[1].set_xlabel("Customers")
axes[1].set_ylabel("Sales")


"""3 Average Sales by Region

Bar plot

X  Region

Y  Average Sales"""
avg_sale = df.groupby('Region')['Sales'].mean().reset_index()
sns.barplot(x='Region',y='Sales',data=avg_sale,ax=axes[2])
axes[2].set_title("Average Sales by Region")
axes[2].set_xlabel("Region")
axes[2].set_ylabel("Sales")
plt.tight_layout()
plt.show()
```
